```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img = Image.open(r'C:\\Users\\DELL\\Downloads\\machinelearn\\kmeans.jpg')
```

```python
img
```

```python
img_array = np.array(img)
```

```python
img_array.shape
```

```python
reshaped_img = img_array.reshape(-1 , img_array.shape[-1])
```

```python
reshaped_img.shape
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters = 10 , init = "k-means++")
```

```python
# When we reshape y:

# clustered_image = y.reshape(img_array.shape[:2])
# This converts the 1D array of labels into a 2D array with the same height and width as the original image:

# Input shape of y: (921600,).
# Output shape after reshaping: (720, 1280).
# Now:

# Each entry in the reshaped array corresponds to the cluster label for the corresponding pixel in the image.

# Why Ignore the 3rd Dimension (RGB)?
# The clustering result (y) is 1 label per pixel, not 1 label per color channel.
# So we only need the height and width to map each label back to its corresponding pixel.
```

```python
y = km.fit_predict(reshaped_img)
y
```

```python
sns.scatterplot(reshaped_img)
```

```python
clustered_image = y.reshape(img_array.shape[:2])

# Plot the original and clustered images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(img)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Clustered Image")
plt.imshow(clustered_image, cmap='viridis')
plt.axis('off')
```
